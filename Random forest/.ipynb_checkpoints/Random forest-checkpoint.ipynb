### Importacion de librerias 

```python
import pandas as pd
from random import sample 
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.metrics import mean_squared_error
```

### Datos input y output

```python
datos = pd.read_csv("Base de datos.csv")
datos.sample(frac=9/10,replace=True)
```

```python
datos.columns[:-1]
sample(list(datos.columns[:-1]),3)
```

```python
bosque = RandomForestRegressor(n_estimators = 500,
                               criterion = "squared_error",
                               max_features = "sqrt",
                               bootstrap = True,
                               max_samples = 9/10,
                               oob_score = True)

bosque.fit(datos[datos.columns[:-1]].values,datos["contraccion"].values)
print(bosque.score(datos[datos.columns[:-1]].values,datos["contraccion"].values))
print(bosque.oob_score_)
```

```python
for arbol in bosque.estimators_:
    tree.plot_tree(arbol,feature_names=list(datos.columns[:-1]))
   # plt.show()
```

```python
predicciones = bosque.predict(datos[datos.columns[:-1]].values)

# Calcular y mostrar el Mean Squared Error (MSE)
mse = mean_squared_error(datos["contraccion"].values, predicciones)
print(f'Mean Squared Error (MSE): {mse}')

# También puedes imprimir otras métricas como el R2 para tener una visión más completa
r2 = bosque.score(datos[datos.columns[:-1]].values, datos["contraccion"].values)
print(f'R^2 Score: {r2}')
```
