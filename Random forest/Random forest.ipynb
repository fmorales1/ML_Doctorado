### Datos input y output

```python
datos = pd.read_csv("Base de datos.csv")
datos.sample(frac=9/10,replace=True)
```

```python
datos.columns[:-1]
sample(list(datos.columns[:-1]),3)
```

```python
bosque = RandomForestRegressor(n_estimators = 100,
                               criterion = "squared_error",
                               max_features = "sqrt",
                               max_depth = None,
                               bootstrap = True,
                               max_samples = 9/10,
                               oob_score = True)

param_grid = {
    "n_estimators": [50, 100, 500],
    "max_depth": [None, 5, 10,15,20],
    "min_samples_split": [2, 5, 10, 15],
    "min_samples_leaf": [1, 2, 4]
}
grid_search = GridSearchCV(bosque, param_grid, cv=5)
grid_search.fit(datos[datos.columns[:-1]].values,datos["contraccion"].values)


```

```python
print(grid_search.best_params_)
```
